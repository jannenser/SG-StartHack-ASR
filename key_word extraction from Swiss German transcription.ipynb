```python
from multi_rake import Rake
import pandas as pd
```

```python
# locally saved file with transcribed Swiss German SG_1-4.xlsx
text = ''
for i in range(1, 5):
    doc_name = 'SG_' + str(i) + '.xlsx'
    #len_sheetnames = len((pd.ExcelFile(doc_name)).sheet_names)
    #print(len_sheetnames)
    for j in range(1, 16):
        sheet_name = 'Anruf (' + str(j) + ')' 
        df = pd.read_excel(doc_name, sheet_name=sheet_name, header=0)
        #print(df[13:].sample(3))
        df = df[13:].fillna('')
        for row in df.iterrows():
            #print(row)
            print(row[1].replace("Unnamed", ""))
            text += row[1]
text = text.replace("NaN", "")
text = text.replace("Unnamed:", "")
#text = text.split('\n').join(' ')
#print(text)
```

```python
text = text[0] + text[1] +text[2]
len(text)
```

```python
rake = Rake(max_words_unknown_lang=2)

keywords = rake.apply(text[1000:1100], text_for_stopwords=text[:1000])

print(keywords)
```
